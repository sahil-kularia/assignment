```python
# Q1 - Part I: Randomized Sales Data Generation (NumPy)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(102316075)  

sales_data = np.random.randint(1000, 5000, size=(12, 4))  # 12 months, 4 categories
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
df = pd.DataFrame(sales_data, columns=["Electronics", "Clothing", "Home & Kitchen", "Sports"], index=months)

# Q1 - Part II: Data Manipulation and Analysis
df["Total Sales"] = df.sum(axis=1)
df["Growth Rate"] = df["Total Sales"].pct_change() * 100

total_sales_category = df.iloc[:, :-2].sum()
total_sales_monthly = df["Total Sales"]

avg_sales_growth = df.iloc[:, :-2].pct_change().mean()

roll_number = 102316075
if roll_number % 2 == 0:
    df["Electronics"] *= 0.90  # Apply 10% discount to Electronics
else:
    df["Clothing"] *= 0.85  # Apply 15% discount to Clothing

# Q1 - Part III: Visualizations
plt.figure(figsize=(12, 6))
df.iloc[:, :-2].plot(marker="o", linestyle="-")
plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(data=df.iloc[:, :-2])
plt.title("Sales Distribution per Category")
plt.xlabel("Product Categories")
plt.ylabel("Sales")
plt.grid(True)
plt.show()

# Q2
array = np.array([[1, -2, 3], [-4, 5, -6]])
absolute_values = np.abs(array)
percentiles = np.percentile(array.flatten(), [25, 50, 75])
percentiles_col = np.percentile(array, [25, 50, 75], axis=0)
percentiles_row = np.percentile(array, [25, 50, 75], axis=1)
mean_val = np.mean(array)
median_val = np.median(array)
std_dev = np.std(array)

# Q3
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])
floor_values = np.floor(a)
ceil_values = np.ceil(a)
trunc_values = np.trunc(a)
rounded_values = np.round(a)

# Q4
def swap_list_elements(lst, idx1, idx2):
    temp = lst[idx1]
    lst[idx1] = lst[idx2]
    lst[idx2] = temp
    return lst

list_example = [1, 2, 3, 4, 5]
swapped_list = swap_list_elements(list_example, 1, 3)

# Q5
def swap_set_elements(s, elem1, elem2):
    s_list = list(s)
    idx1, idx2 = s_list.index(elem1), s_list.index(elem2)
    s_list[idx1], s_list[idx2] = s_list[idx2], s_list[idx1]
    return set(s_list)

set_example = {10, 20, 30, 40}
swapped_set = swap_set_elements(set_example, 20, 40)
```
